# 17.1 Interpolation Problem Statement
Interpolation is used when we have a data set with independent values (x) and dependent values (y). Interpolation is used to estimate the dependent value based on an independent value that is not part of the data set, but between values in the data set. The interpolation function is often called y_hat.

## 17.2 Linear Interpolation
Linear interpolation is the most basic type. It assumes that an estimated point falls on a straight line between the two points on either side.



```python
# for the data x = [1, 2, 3] and y = [2, 3, 4], interpolate for x = 2.5.
3 + (4-3)*(2.5-2)/(3-2)
```

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 3, 4]

f = interp1d(x, y)
y_hat = f(2.5)
print(y_hat)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

x = [1, 2, 3, 4, 5, 6]
y = [4, 7, 2, 8, 7, 5]
x_new = np.linspace(1,6,1000)
f = interp1d(x,y)
y_new = f(x_new)
plt.figure(figsize = (10,8))
plt.plot(x, y, 'ro')
plt.plot(x_new,y_new, 'b')
plt.title('Linear Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Cubic Spline Interpolation
Cubic spline interpolation uses a set of piecewise cubic functions to estimate unknown values. n points require n-1 cubic functions to interpolate, and each of these functions has four coefficients.
The cubic function endpoints must be the data points on either side, and the first and second derivatives must be continuous at these data points to ensure smoothness. The function at the first and last data point in the set have more freedom, but a common approach is to set the endpoint derivative to zero.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [4, 7, 2, 8, 7, 5]

f = CubicSpline(x, y, bc_type='natural')
# bc_type = 'natural' uses the constraints explained above
x_new = np.linspace(1,6,1000)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```
